```python
import pandas as pd
global_emissions = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions')
global_emissions [1]



```

```python
global_emissions[1].columns = ['Landen','uitstoot 1990','uitstoot 2005','uitstoot 2017','2017 of world','2017vs1990','2017 per land','2017 per capita','2018 tot inc lucf','2018 tot ex lucf']
global_emissions[1]
```

```python
gefilterde_tabel = global_emissions[1].drop(global_emissions[1].index[[0,1,2,67]])
gefilterde_tabel
```

```python
gesorteerde_tabel = gefilterde_tabel.sort_values(by='uitstoot 2017',axis=0,ascending=False,inplace=False,kind='quicksort',na_position='last',ignore_index=True).head(5)
gesorteerde_tabel
```

```python
gesorteerde_tabel[['Landen','uitstoot 1990','uitstoot 2005','uitstoot 2017']]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
landen = gesorteerde_tabel.loc[:,'Landen']
uitstoot_1990 = gesorteerde_tabel.loc[:,'uitstoot 1990']
uitstoot_2005 = gesorteerde_tabel.loc[:,'uitstoot 2005']
uitstoot_2017 = gesorteerde_tabel.loc[:,'uitstoot 2017']
ax.plot(landen,uitstoot_1990, label='uitstoot 1990')
ax.plot(landen,uitstoot_2005, label='uitstoot 2005')
ax.plot(landen,uitstoot_2017, label='uitstoot 2017')
plt.title('top 5 landen met de grootste uitstoot CO2')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
index = 'uitstoot 1990','uitstoot 2005','uitstoot 2017'
uitstoot_china = gesorteerde_tabel.loc[0,['uitstoot 1990','uitstoot 2005','uitstoot 2017']]
uitstoot_unitedstates = gesorteerde_tabel.loc[1,['uitstoot 1990','uitstoot 2005','uitstoot 2017']]
uitstoot_india = gesorteerde_tabel.loc[2,['uitstoot 1990','uitstoot 2005','uitstoot 2017']]
uitstoot_rusland = gesorteerde_tabel.loc[3,['uitstoot 1990','uitstoot 2005','uitstoot 2017']]
uitstoot_japan = gesorteerde_tabel.loc[4,['uitstoot 1990','uitstoot 2005','uitstoot 2017']]
ax.plot(index,uitstoot_china, label='china')
ax.plot(index,uitstoot_unitedstates, label='united states')
ax.plot(index,uitstoot_india, label='india')
ax.plot(index,uitstoot_rusland, label='rusland')
ax.plot(index,uitstoot_japan, label='japan')
plt.title('top 5 landen met de grootste uitstoot CO2')
ax.legend()
plt.show()
```
