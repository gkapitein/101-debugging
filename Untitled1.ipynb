```python
import pandas as pd
global_emissions = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions')
global_emissions [1]

```

```python
global_emissions[1].columns = ['Landen','uitstoot 1990','uitstoot 2005','uitstoot 2017','2017 of world','2017vs1990','2017 per land','2017 per capita','2018 tot inc lucf','2018 tot ex lucf']
global_emissions[1]
```

```python
gefilterde_tabel = global_emissions[1].drop(global_emissions[1].index[[0,1,2,67]])
gefilterde_tabel
```

```python
landen = gefilterde_tabel.loc[:,['Landen','uitstoot 1990','uitstoot 2005','uitstoot 2017']]
landen
```

```python
landen[landen['uitstoot 1990']>5]
```

```python
landen['change uitstoot 1990']=landen.loc[:,'uitstoot 1990']/landen.loc[:,'uitstoot 1990']*100
landen
```

```python
landen['change uitstoot 2005']=landen.loc[:,'uitstoot 2005']/landen.loc[:,'uitstoot 1990']*100
landen
```

```python
landen['change uitstoot 2017']=landen.loc[:,'uitstoot 2017']/landen.loc[:,'uitstoot 2005']*100
landen
```

```python
landen_gesorteerd = landen.sort_values(by='change uitstoot 2017',axis=0,ascending=False,inplace=False,kind='quicksort',na_position='last',ignore_index=True)
landen_gesorteerd
```

```python
eerste3 = landen_gesorteerd.head(3)
eerste3
```

```python
laatste3 = landen_gesorteerd.tail(3)
laatste3
```

```python
eerstenlaatste3=pd.concat([eerste3,laatste3])
eerstenlaatste3
```

```python
eerstenlaatste3.loc[:,['Landen','change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
index = 'relatieve uitstoot 1990','relatieve uitstoot 2005','relatieve uitstoot 2017'
uitstoot_tencislands = eerstenlaatste3.loc[0,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
uitstoot_afghanistan = eerstenlaatste3.loc[1,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
uitstoot_bvislands = eerstenlaatste3.loc[2,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
uitstoot_ukraine = eerstenlaatste3.loc[205,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
uitstoot_yemen = eerstenlaatste3.loc[206,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
uitstoot_noordkorea = eerstenlaatste3.loc[207,['change uitstoot 1990','change uitstoot 2005','change uitstoot 2017']]
ax.plot(index,uitstoot_tencislands, label='Turks and Caicos Islands')
ax.plot(index,uitstoot_afghanistan, label='Afghanistan')
ax.plot(index,uitstoot_bvislands, label='British Virgin Islands')
ax.plot(index,uitstoot_ukraine, label='Ukraine')
ax.plot(index,uitstoot_yemen, label='Yemen')
ax.plot(index,uitstoot_noordkorea, label='Noord Korea')
plt.title('3 grootste landen en 3 laagste landen met relatieve CO2 uitstoot')
ax.legend()
plt.show()
```
